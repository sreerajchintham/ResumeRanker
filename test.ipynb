```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
df.info()
```

```python
# Select variables to visualize
columns_to_plot = ["Glucose", "BloodPressure", "BMI", "Age"]

```

```python
# Create histograms with KDE plots
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30, color="blue", alpha=0.7)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")

plt.tight_layout()
plt.show()

```

```python
# Create box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[columns_to_plot], palette="Set2")
plt.xlabel("Variables")
plt.ylabel("Values")
plt.title("Boxplot of Selected Variables")
plt.show()
```
